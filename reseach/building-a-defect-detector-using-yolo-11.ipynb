```python
import yaml
dataset_name="neu_dataset_yolo_ready"
dataset_path=f"/kaggle/input/neu-dataset-prepared-for-yolo/{dataset_name}"
```

```python
with open(f"{dataset_path}/data.yaml","r") as f :
     original_data=yaml.safe_load(f)
```

```python
new_data={
    "train":f"{dataset_path}/train/images",
    "val":f"{dataset_path}/val/images",
    "nc":original_data["nc"],
    "names":original_data["names"]
}
```

```python
output_path=f"/kaggle/working/data.yaml"
with open(output_path,"w") as f:
    yaml.dump(new_data,f)
```

```python
with open(output_path,"r") as f:
    print(f.read())
```

```python
!yolo task=detect mode=train data={output_path} model="yolo11n.pt" epochs=50 imgsz=640
```
